### 1. function definition

#### semantic_substitution: antonym 

```python
import nltk
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize
from nltk import pos_tag
import random

nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

def generate_sample_dict(zh, en_ori, en_modi, modi_type, ori_word, modi_word):
    return {"zh_sent": zh,
            "en_ori_sent": en_ori,
            "en_modi_sent": en_modi,
            "modi_type": modi_type, # ["grammer_swap", "grammer_agreement", "grammer_punc", "semantic_substitution", "semantic_omission", "semantic_addition", "semantic_ambiguity"]
            "ori_word": ori_word,
            "modi_word": modi_word}


def smart_join(wordss):
    flag = False
    sentence = ""
    for word in wordss:
        if word in [".", ",", "!", "?", ";", ":"]:
            sentence += word
        elif word in ["``"]:
            sentence += '''"'''
            flag=True
        elif word in ["''"]:
            sentence += '''"'''
        else:
            if flag:
                sentence += word
                flag=False
            else:
                sentence += " " + word
            
    return sentence.lstrip()  # remove leading space


def get_antonyms(word):
    antonyms = []
    for syn in wordnet.synsets(word):
        for lemma in syn.lemmas():
            if lemma.antonyms():
                antonyms.append(lemma.antonyms()[0].name())
    return antonyms

def replace_random_noun_with_antonym(sentence):
    wordss = word_tokenize(sentence)
    pos_tags = pos_tag(wordss)

    # Get the nouns in the sentence
    nouns = [word for word, pos in pos_tags if pos in ['NN', 'NNS', 'NNP', 'NNPS']]

    if not nouns:
        return sentence  # return the original sentence if no noun is found

    # Randomly select a noun to replace
    noun_to_replace = random.choice(nouns)

    # Get the antonyms of the selected noun
    antonyms = get_antonyms(noun_to_replace)
    
    if not antonyms:
        for n in nouns:
            antonyms = get_antonyms(n)
            if antonyms:
                break

    if not antonyms:
        return sentence, "NA", "NA"  # return the original sentence if no antonym is found

    # Replace the selected noun with its first antonym
    new_words = [antonyms[0] if word == noun_to_replace else word for word in wordss]

    return smart_join(new_words), noun_to_replace, antonyms[0]

sentence = "The cat is playing with a ball, and I am watching the dog."
new_sentence, ori_word, modi_word = replace_random_noun_with_antonym(sentence)
print(new_sentence, ori_word, modi_word)
```

#### semantic addition / semantic_omission 

```python
import random
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import words

nltk.download('words')

def add_word_at_random_position(sentence):
    wordss = word_tokenize(sentence)
    position = random.randint(0, len(wordss))
    word_to_add = random.choice(words.words()) 
    wordss.insert(position, word_to_add)
    return smart_join(wordss), word_to_add, word_to_add

# delete word at random
# def delete_word_at_random_position(sentence):
#     wordss = word_tokenize(sentence)
#     if len(wordss) > 1:  # ensure there's at least one word left
#         position = random.randint(0, len(wordss) - 1)
        
#         if position == len(wordss)-1:
#             left = wordss[position-1]
#             right = wordss[position-1]
            
#         elif position == 0:
#             left = wordss[position+1]
#             right = wordss[position+1]
            
#         else:
#             left = wordss[position-1]
#             right = wordss[position+1]

#         del wordss[position]
#     return smart_join(wordss), left, right

# delete noun or verb at random

def delete_word_at_random_position(sentence):
    wordss = word_tokenize(sentence)
    if len(wordss) > 1:  # ensure there's at least one word left
        position = random.randint(0, len(wordss) - 1)
        
        pos_tags = pos_tag(wordss)

        # Get the nouns in the sentence
        positions = [idx for idx, (word, pos) in enumerate(pos_tags) if pos in ['NN', 'NNS', 'NNP', 'NNPS', 'VB', 'VBP', 'VBZ']]
        position = random.choice(positions)
        
        if position == len(wordss)-1:
            left = wordss[position-1]
            right = wordss[position-1]
            
        elif position == 0:
            left = wordss[position+1]
            right = wordss[position+1]
            
        else:
            left = wordss[position-1]
            right = wordss[position+1]

        del wordss[position]
    return smart_join(wordss), left, right

sentence = '''"The cat is playing with a ball, and I am watching the dog."'''
new_sentence, ori_word, modi_word = add_word_at_random_position(sentence)
print(new_sentence, ori_word, modi_word)

new_sentence, ori_word, modi_word = delete_word_at_random_position(sentence)
print(new_sentence, ori_word, modi_word)
```

## Syntax Mutation

### 1. function definition

#### words swapping

```python
def swap_two_random_words(sentence):
    puncs = [".", ",", "!", "?", ";", ":", '''"''', "``", "`", "''", "'"]
    wordss = word_tokenize(sentence)
    if len(wordss) > 2:  # ensure there are at least two words to swap
        position1 = random.randint(0, len(wordss) - 2)
        while wordss[position1] in puncs:
            position1 = random.randint(0, len(wordss) - 2)
        position2 = position1
        while position2 == position1 or position2 in puncs:  # ensure we get two different positions
            position2 = random.randint(0, len(wordss) - 2)
        wordss[position1], wordss[position2] = wordss[position2], wordss[position1]

    elif len(wordss) == 2:  # ensure there are at least two words to swap
        position1 = random.randint(0, len(wordss) - 1)
        while wordss[position1] in puncs:
            position1 = random.randint(0, len(wordss) - 1)
        position2 = position1
        while position2 == position1 or position2 in puncs:  # ensure we get two different positions
            position2 = random.randint(0, len(wordss) - 1)
        wordss[position1], wordss[position2] = wordss[position2], wordss[position1]
        
    return smart_join(wordss), wordss[position1], wordss[position2]

# example
sentence = '''The cat is playing with a ball, and I am watching the dog!'''
new_sentence, w1, w2 = swap_two_random_words(sentence)
print(new_sentence, w1, w2)
```

#### subject-verb agreement

```python
from nltk.stem import WordNetLemmatizer

def break_subject_verb_agreement(sentence):
    wordss = word_tokenize(sentence)
    pos_tags = pos_tag(wordss)

    lemmatizer = WordNetLemmatizer()

    # Get the verbs in the sentence
    verbs = [word for word, pos in pos_tags if pos in ['VB', 'VBP', 'VBZ']]

    if not verbs:
        return sentence, "NA", "NA"  # return the original sentence if no verb is found

    # Select a random verb to change
    verb_to_change = random.choice(verbs)
    
    if verb_to_change[-2:]=="es" and len(verbs) ==1:
        return sentence, "NA", "NA"
    
    while verb_to_change[-2:]=="es" and len(verbs)>1:
        verb_to_change = random.choice(verbs)
    
    
    if verb_to_change in ["am", "is", "are"]:
        for v in verbs:
            if v not in ["am", "is", "are"] and v[-2:]!="es":
                verb_to_change = v
    
    if verb_to_change in ["am", "is", "are"] or verb_to_change[-2:]=="es":
        return sentence, "NA", "NA"

    # Change the verb form to break the agreement
    if verb_to_change.endswith('s'):
        new_verb = lemmatizer.lemmatize(verb_to_change, 'v')
    else:
        new_verb = verb_to_change + 's'

    # Replace the selected verb with the new verb
    new_words = [new_verb if word == verb_to_change else word for word in wordss]

    return smart_join(new_words), verb_to_change, new_verb

sentence = "In case of more severe disturbances, there are also fluctuations in the electricity grid. Satellites providing navigation information are also vulnerable."
new_sentence, w1, w2 = break_subject_verb_agreement(sentence)
print(new_sentence, w1, w2)
```

#### punctuation replacement

```python
def change_random_punctuation(sentence):
    indices = [i for i, char in enumerate(sentence) if char in ['!', '?']]
    if not len(indices):
        return sentence, "NA", "NA"
    if indices:
        index_to_change = random.choice(indices)
        sentence = list(sentence)  # convert string to list to allow item assignment
        original_punc = sentence[index_to_change]
        sentence[index_to_change] = '!' if sentence[index_to_change] == '?' else '?'
        sentence = ''.join(sentence)  # convert list back to string
    return sentence, original_punc, sentence[index_to_change]

# example
sentence = '''"The cat is playing with a ball, and I am watching the dog!"'''
new_sentence, w1, w2 = change_random_punctuation(sentence)
print(new_sentence, w1, w2)
```

### 2. apply the syntax mutation to 200 samples (100 for original, 100 for llm)

```python
ori_data = load_csv('sample_200_for_gold.csv')
llm_data = load_csv('data_from_llm.csv')

ori4syntax = ori_data[:100]
llm4syntax = llm_data[:100]

modi_type_list = ["grammer_swap", "grammer_agreement", "grammer_punc"]

result = []

for idx, instance in ori4syntax.iterrows():

    en_data = instance['en']
    zh_data = instance['zh']
    
    modi_type = random.choice(modi_type_list)
    if modi_type == "grammer_swap":
        new_sentence, ori_word, modi_word = swap_two_random_words(en_data)
    elif modi_type == "grammer_agreement":
        new_sentence, ori_word, modi_word = break_subject_verb_agreement(en_data)
    elif modi_type == "grammer_punc":
        new_sentence, ori_word, modi_word = change_random_punctuation(en_data)
        
    if ori_word=="NA":
        modi_type = random.choice(["grammer_swap"])
        if modi_type == "grammer_swap":
            new_sentence, ori_word, modi_word = swap_two_random_words(en_data)

            
    result.append(generate_sample_dict(zh_data, en_data, new_sentence, modi_type, ori_word, modi_word))

save_json(result, "ori_syntax_0-100.json")


result = []

for idx, instance in llm4syntax.iterrows():

    en_data = instance['en']
    zh_data = instance['zh']
    
    modi_type = random.choice(modi_type_list)
    if modi_type == "grammer_swap":
        new_sentence, ori_word, modi_word = swap_two_random_words(en_data)
    elif modi_type == "grammer_agreement":
        new_sentence, ori_word, modi_word = break_subject_verb_agreement(en_data)
    elif modi_type == "grammer_punc":
        new_sentence, ori_word, modi_word = change_random_punctuation(en_data)
        
    if ori_word=="NA":
        modi_type = "grammer_swap"
        if modi_type == "grammer_swap":
            new_sentence, ori_word, modi_word = swap_two_random_words(en_data)
            
    result.append(generate_sample_dict(zh_data, en_data, new_sentence, modi_type, ori_word, modi_word))

save_json(result, "llm_syntax_0-100.json")
```
